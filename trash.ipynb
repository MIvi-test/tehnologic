```python
import requests
import pandas as pd
```

```python
r = requests.get("https://dfedorov.spb.ru/python3/sport.txt")
r.encoding = 'cp1251'
r = r.text
```

```python
col = r.split('\n')[0].split('\t') # pyright: ignore[reportAttributeAccessIssue]
d = []
for string in r.split('\n')[1:-1]: # type: ignore
    words = string.split('\t')
    d.append(tuple(words[1:]))
    
df = pd.DataFrame(d, columns= col[1:])
```

```python
from collections import defaultdict
ves = defaultdict(int)
for i in df.values:
    data = i[2]
    m = map(str.strip, data.split(','))
    for key in m:
        if key:
            ves[key] += 1
result = sorted(ves, key = ves.get, reverse=True)
```
