```python
class MethodError(Exception):
    def __init__(self):
        super().__init__('данная функция недоступна для данного класса')
```

```python
class MyFile:
    def __init__(self, path: str, method: str):
        if method not in ["read", "write", "append", "url"]:
            raise ValueError
        if method == 'url' and 'https://' not in path:
            raise ValueError
        if method == 'url':
            from requests import get
            self.get = get
        self.path = path
        self.method = method

    def read(self) -> str:
        if self.method != 'read':
            raise MethodError
        with open(self.path, 'r') as f:
            text = f.read()
        return text

    def write(self, text: str) -> str:
        if self.method != 'write':
            raise MethodError
        with open(self.path, 'w') as f:
            f.write(text)
        return text

    def append(self, text: str) -> str:
        if self.method != 'append':
            raise MethodError
        with open(self.path, 'a') as f:
            f.write(text)
        return text

    def read_url(self) -> str:
        if self.method != 'url':
            raise MethodError

        return self.get(self.path).text

    def count_urls(self) -> int:
        if self.method != 'url':
            raise MethodError
        return self.get(self.path).text.count("https://")

    def write_url(self, filename) -> None:
        if self.method != 'url':
            raise MethodError
        with open(filename, 'a') as f:
            f.write(self.get(self.path).text)
        return
```
